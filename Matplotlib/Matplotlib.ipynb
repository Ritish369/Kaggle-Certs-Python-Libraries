Matplotlib is like the mother of all data visualization libraries in python.

Understanding the type of data one Is dealing With Is important In data visualization.

Univariate Analysis = When a single column is plotted.

Bivariate Analysis = When 2 columns are plotted together in a graph.

Multivariate Analysis = When more than 2 columns are plotted together.

```python
# Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')

print("Setup complete.")
```

### 2D Line plot

Used for bivariate analysis. On categorical - numerical and numerical - numerical data type columns.
Generally, 2D plots are plotted on Time Seires data.

```python
# Plotting a simple function
price = [48000, 54000, 57000, 49000, 47000, 45000]
year = [2015, 2016, 2017, 2018, 2019, 2020]
# plot(x-axis, y-axis) -> plots a line plot; Categorical column on x and numerical column on y
plt.plot(year, price)
```

```python
# Form a pandas dataframe
# IPL dataset
batsman = pd.read_csv(r'C:\Users\Ritish Garg\Desktop\sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# colors(hex codes) and line(width and style) and marker(size){Markers are the points which are joined to form line plots}
# linestyle = solid, dashed, dotted, 'dashed dot'
# marker = 'D', '+', '.'. '>', 'o', '<' and more
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = 3, marker = 'D', markersize = 10)
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#000000', linestyle = 'dashed', linewidth = 2, marker = 'o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend(table that's on the graph, kinda scale table) -> location => label parameter & legend() function
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = 3, marker = 'D', markersize = 10, label = 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#000000', linestyle = 'dashed', linewidth = 2, marker = 'o', label = 'Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# loc positions the legend table at best empty place possible & many options are there to assign it
plt.legend(loc = 'best')
```

```python
# Limiting axes {Generally, it is done when some outliers are present and the other points' plot/pattern disappears, in some sense.}
price = [48000, 54000, 57000, 49000, 47000, 45000, 4500000]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)

#To trim the y-axis values to avoid any kind of outliers and bring it into a range, in some sense.
plt.ylim(0, 75000) # y limit
# Similar thing can be done on x-axis
plt.xlim(2017, 2019) # x limit
```

```python
# Grid and Show
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = 3, marker = 'D', markersize = 10, label = 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#000000', linestyle = 'dashed', linewidth = 2, marker = 'o', label = 'Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# Used to get better readability out of the plot
plt.grid()

# It is the main fn due to which graphs get plotted. Also, when coding is done in proper text editor, graphs are not plotted.
# Then, in tthat case, use show() function.
plt.show()
```

### Bar Chart

Categorical data/column on X-axis and an aggregate operation is performed on Y-axis.

Used for both univariate and bivariate analysis, in some sense.
Generally, on numerical vs categorical types of data, where also the numerical one is the aggregate operation. Its usecase is in aggregate analysis of groups.

```python
# Simple bar chart, color
children = [10, 20, 40, 10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']

plt.bar(colors, children, color = 'black')

# Labels as for others are the same way.
```

```python
# Horizontal bar chart(Should be used when the number of categories is large)
plt.barh(colors, children, color = 'black')
```

```python
# Multiple Bar Charts (Side-by-side bar plots i.e., different values for different things of the same thing.)
df = pd.read_csv(r"C:\Users\Ritish Garg\Desktop\batsman_season_record.csv")
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width = 0.2, color = 'yellow') # width parameter is for the bar/bin width
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2, color = 'red')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2, color = 'blue')

# xticks() function is used to re-change the x-axis values from changed to the original.
plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# Problem (X-axis values for categorical data overlap each other since they have long names or there are too many categories.)
children = [10, 20, 40, 10, 30]
colors = ['red red red red red red', 'blue blue blue blue', 'green green green green green', 'yellow yellow yellow yellow', 'pink pink pink pink']

plt.bar(colors, children, color = 'black')

# Solution to problem(Displays all the names now vertically)
plt.xticks(rotation = 'vertical')

plt.show()
```

```python
# Stacked Bar Chart( done using a bottom parameter of bar() function)
plt.bar(df['batsman'], df['2017'], color = 'red', label = '2017')
plt.bar(df['batsman'], df['2016'], bottom = df['2017'], color = 'yellow', label = '2016')
plt.bar(df['batsman'], df['2015'], bottom = (df['2016'] + df['2017']), color = 'orange', label = '2015')

plt.legend()
plt.show()
```

### Subplots

Multiple graphs can be plotted simultaneously. This is required when dealing with multiple columns and want to plot side-by-side graphs together instead of individual graphs. Done to see what's happening b/w those columns.

```python
# A diff way to plot graphs
batters.head()
```

```python
# Way done till now
plt.figure(figsize = (15, 6))

plt.scatter(batters['avg'], batters['strike_rate'])

plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
# This is the diff way which will also be used while plotting subplots.
# figure object and axis object are obtained on calling subplots() fn
fig, ax = plt.subplots(figsize = (15, 6))

ax.scatter(batters['avg'], batters['strike_rate'], color = 'red', marker = '+')

ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

This above diff way approach is used because one can plot side-by-side multiple graphs.

```python
# Now, plotting 2 side-by-side plots -- runs vs avg and avg vs strike_rate -- using subplots()
# No. of graphs plotted = nrows * ncols
# sharex parameter tells to share the x-axis for the graphs.
fig, ax = plt.subplots(nrows = 2, ncols = 1, sharex = True, figsize = (10, 6))

ax[0].scatter(batters['avg'], batters['strike_rate'], color = 'red')
ax[1].scatter(batters['avg'], batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10))

# As ax is now a 2 x 2 matrix of graphs
ax[0, 0].scatter(batters['avg'], batters['strike_rate'], color = 'red')
ax[0, 1].scatter(batters['avg'], batters['runs'])
ax[1, 0].hist(batters['avg'])
ax[1, 1].hist(batters['runs'])
```

```python
# Another (third) way for plotting graphs than the different approach above
fig = plt.figure()

# add_subplot(row number, col number, graph's position/number in the subplot graph)
# Add as many axis as you want like axi, i = 1, 2, 3,.........
ax1 = fig.add_subplot(2, 2, 1)
ax1.scatter(batters['avg'], batters['strike_rate'], color = 'red')

ax2 = fig.add_subplot(2, 2, 2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2, 2, 3)
ax3.hist(batters['avg'])

# This is a hectic way. So, go with the second above approach.
```

### 3D Surface Plots

Used very much in Machine Learning, especially in plotting the loss functions and all.

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
```

```python
# Forms a grid
xx, yy = np.meshgrid(x, y)
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize = (12, 8))

ax = plt.subplot(projection = '3d')

# Main function that plots the 3d surfaces
graph = ax.plot_surface(xx, yy, z, cmap = 'viridis')
fig.colorbar(graph)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (12, 8))

ax = plt.subplot(projection = '3d')

graph = ax.plot_surface(xx, yy, z, cmap = 'viridis')
fig.colorbar(graph)
```

### Contour Plots

Contour is a special plot which represents the 3D graph into a 2D graph.

```python
fig = plt.figure(figsize = (12, 8))

ax = plt.subplot()

graph = ax.contour(xx, yy, z, cmap = 'viridis')
fig.colorbar(graph)
```

```python
# Another contour plot variation is contourf() i.e., contour fill; its brother since its easier to understand
fig = plt.figure(figsize = (12, 8))

ax = plt.subplot()

graph = ax.contourf(xx, yy, z, cmap = 'viridis')
fig.colorbar(graph)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (12, 8))

ax = plt.subplot()

graph = ax.contourf(xx, yy, z, cmap = 'viridis')
fig.colorbar(graph)
```
